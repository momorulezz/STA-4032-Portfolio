```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming x, y, and y2 are defined
# For example:
# x = np.linspace(-10, 10, 100)
# y = 2 * x + 1  # True line
# y2 = y + np.random.normal(0, 2, x.size)  # Adding some noise

plt.figure(figsize=(10, 6))  # Initialize the plot outside the loop

for m_guess in np.linspace(-100, 100, 100):
    y_guess = m_guess * x
    loss = np.sum((y_guess - y)**2)
    print(loss, 'M = ', m_guess)

# Plotting outside the loop to avoid clutter
plt.plot(x, y_guess, 'r.', label='Last mx+b (guess)')
plt.plot(x, y2, 'g.', label='mx + b + r')
plt.legend(loc='best')
plt.show()
```

```python
#code runs by intializing values for m and b, then declaring what the "loss" variable means, then running an if loop to determine what value to output.
best_loss = 100000000000
best_m = 0
best_b = 0

for m_guess in np.linspace(-100,100,100):
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:

            best_loss = loss
            best_m = m_guess
            best_b = b_guess
```

```python
y_guess = best_m * x + best_b

plt.plot(x,y_guess,'r.', label= 'mx+b(guess)');
plt.plot(x,y2,'g.', label= 'mx+b+r');
plt.legend(loc='best')
plt.show()

print(best_loss,best_m,best_b)
```

```python
#importing data from above spreadsheet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name):
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```
